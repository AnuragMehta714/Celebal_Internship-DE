## Step 1: Load the CSV Dataset into a Pandas DataFrame

```python
import pandas as pd

# Load the dataset
df = pd.read_csv(r'F:\CSE\DOCUMENTS\PYTHON\CSV Data\Combined_dataset.csv')

print('Dataset loaded successfully!')
```

## Step 3: Handle Missing Values

```python
# Identify missing values in each column
print('=== MISSING VALUES PER COLUMN ===')
missing = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_percent})
print(missing_df[missing_df['Missing Count'] > 0])
```

```python
# Fill missing values in 'discount' column with median
median_discount = df['discount'].median()
df['discount'].fillna(median_discount, inplace=True)

# Fill missing 'seller_name' with 'Unknown'
df['seller_name'].fillna('Unknown', inplace=True)

# Drop rows where 'rating' is 0 (invalid rating)
df = df[df['rating'] > 0]

print(f'After cleaning missing values:')
print(f'Remaining missing values: {df.isnull().sum().sum()}')
print(f'Rows remaining: {len(df)}')
```

## Step 4: Basic Operations - Filter Rows & Select Columns

```python
# Filter: Products with rating >= 4.0
high_rated = df[df['rating'] >= 4.0]
print(f'Products with rating >= 4.0: {len(high_rated)}')

# Filter: Products with discount > 50%
high_discount = df[df['discount'] > 50]
print(f'Products with discount > 50%: {len(high_discount)}')
```

```python
# Select only important columns
selected_cols = df[['product_id', 'title', 'product_description',
                     'rating', 'ratings_count', 'initial_price',
                     'discount', 'category']]
print('SELECTED COLUMNS')
print(selected_cols.head())
```

```python
# Group by category and count products
print('PRODUCTS PER CATEGORY')
category_count = df.groupby('category')['product_id'].count().sort_values(ascending=False)
print(category_count)
```

## Step 5: Remove Duplicates

```python
print(f'Before removing duplicates: {len(df)} rows')
before = len(df)

# Remove duplicate rows based on product_id
df = df.drop_duplicates(subset='product_id')

after = len(df)
print(f'After removing duplicates : {after} rows')
print(f'Duplicates removed        : {before - after}')
```

## Step 6: Create Derived Column — `total_amount = price × quantity`


```python
# Assign a default quantity of 1 (as dataset has no quantity column)
df['quantity'] = 1

# total_amount = initial_price * quantity  (as per assignment formula)
df['total_amount'] = df['initial_price'] * df['quantity']

# Bonus: also compute actual discount amount
df['discount_amount'] = df['initial_price'] * (df['discount'] / 100)
df['discount_amount'] = df['discount_amount'].round(2)

print('=== NEW DERIVED COLUMNS ===')
print(df[['title', 'initial_price', 'quantity', 'total_amount', 'discount', 'discount_amount']].head(5))
```

## Step 7: Save the Cleaned Dataset as a New CSV File

```python
# Keep only useful columns for the cleaned dataset
clean_cols = ['product_id', 'title', 'product_description', 'rating',
              'ratings_count', 'initial_price', 'discount', 'discount_amount',
              'quantity', 'total_amount', 'currency', 'category', 'seller_name']

df_clean = df[clean_cols].copy()

# Save to new CSV file
df_clean.to_csv('cleaned_shopping_dataset.csv', index=False)

print('Cleaned dataset saved as: cleaned_shopping_dataset.csv')
print(f'Final Shape: {df_clean.shape[0]} rows × {df_clean.shape[1]} columns')
```
